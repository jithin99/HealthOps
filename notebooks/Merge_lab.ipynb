```python
import pandas as pd
import numpy as np
```

```python
labevents = pd.read_parquet("/Volumes/hc/healthcare/healthcare/silver/labevents")
labitems = pd.read_parquet("/Volumes/hc/healthcare/healthcare/silver/labitems")
```

```python
merge = labevents.merge(
    labitems[['itemid', 'label']],
    on='itemid',
    how='inner'
)
```

```python
merge = merge[merge['label'].str.lower().isin(['glucose', 'creatinine'])]
```

```python
agg_df = merge.groupby(['subject_id', 'hadm_id', 'label'])['valuenum'].mean().reset_index()
```

```python
agg_pivot = agg_df.pivot(index=['subject_id', 'hadm_id'], columns='label', values='valuenum').reset_index()
```

```python
agg_pivot.rename(columns={
    'Glucose': 'avg_glucose',
    'Creatinine': 'avg_creatinine'
}, inplace=True)
```

```python
agg_pivot['hadm_id'] = agg_pivot['hadm_id'].astype('int64')
```

```python
agg_pivot.to_parquet("/Volumes/hc/healthcare/healthcare/silver/agg_pivot")
```
