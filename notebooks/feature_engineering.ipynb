```python
import pandas as pd
```

```python
df = pd.read_parquet("/Volumes/hc/healthcare/healthcare/silver/merged_all")
```

```python
df['risk_score'] = (
    0.4 * df['avg_glucose'] +
    0.3 * df['avg_creatinine'] +
    0.3 * df['length_of_stay']
)
```

```python
def risk_category(x):
    if x > 120:
        return 'High'
    elif x > 80:
        return 'Medium'
    else:
        return 'Low'
```

```python
df['risk_category'] = df['risk_score'].apply(risk_category)
```

```python
df.to_parquet("/Volumes/hc/healthcare/healthcare/gold/patient_features")
```
