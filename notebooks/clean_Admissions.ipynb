```python
import pandas as pd
```

```python
dbutils.fs.mkdirs("/Volumes/hc/healthcare/healthcare/silver")
dbutils.fs.mkdirs("/Volumes/hc/healthcare/healthcare/gold")
```

```python
df=pd.read_csv('/Volumes/hc/datasets/raw/ADMISSIONS.csv')
df.info()
```

```python
df['admittime'] = pd.to_datetime(df['admittime'])
df['dischtime'] = pd.to_datetime(df['dischtime'])
```

```python
df['length_of_stay'] = (df['dischtime'] - df['admittime']).dt.days
```

```python
df.drop(columns=[
    'row_id',
    'edregtime',
    'edouttime',
    'has_chartevents_data',
    'language',
    'religion'
], inplace=True)
```

```python
df.drop(columns=['deathtime'],inplace=True)
```

```python
df['admission_type'] = df['admission_type'].str.strip().str.upper()
```

```python
val = {
    'EMERGENCY': 2,
    'URGENT': 1,
    'ELECTIVE': 0
}

df['emergency_flag'] = df['admission_type'].map(val)
```

```python
df.to_parquet("/Volumes/hc/healthcare/healthcare/silver/admissions")
```
