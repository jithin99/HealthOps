```python
import pandas as pd
```

```python
admissions = pd.read_parquet("/Volumes/hc/healthcare/healthcare/silver/admissions")
patients = pd.read_parquet("/Volumes/hc/healthcare/healthcare/silver/patients")
agg_pivot = pd.read_parquet("/Volumes/hc/healthcare/healthcare/silver/agg_pivot")
```

```python
df = admissions.merge(
    agg_pivot,
    on=['subject_id', 'hadm_id'],
    how='left'
)
```

```python
df = df.merge(
    patients,
    on='subject_id',
    how='left'
)
```

```python
df.to_parquet("/Volumes/hc/healthcare/healthcare/silver/merged_all")
```
