```python
import pandas as pd
```

```python
df = pd.read_parquet("/Volumes/hc/healthcare/healthcare/gold/patient_features")
```

```python
spark_df = spark.createDataFrame(df)

spark_df.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("hc.final_data.patient_data")
```

```python
risk_df = spark_df.select(
    "subject_id",
    "hadm_id",
    "avg_glucose",
    "avg_creatinine",
    "length_of_stay",
    "risk_score",
    "risk_category"
)

risk_df.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("hc.final_data.patient_risk_scores")
```
